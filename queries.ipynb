```python
#12. Filter by Road Condition

road_options = df['Road Condition'].dropna().unique()
print("\033[1;31mSelect Road Condition:\033[0m")
for i, condition in enumerate(road_options):
    print(f"{i+1}. {condition}")

choice = int(input("\033[1;31mEnter your choice number: \033[0m"))
selected_road = road_options[choice - 1]

result = df[df['Road Condition'] == selected_road]
print(f"\033[1;31mTotal accidents on {selected_road} roads:\033[0m \033[1;33m{len(result)}\033[0m")
```

```python
#13 Find Date with Most Accidents in Selected Location

locations = df['Location'].dropna().unique()
print("\033[1;31mChoose a location:\033[0m")
for i, loc in enumerate(locations): print(f"{i+1}. {loc}")
choice = int(input("\033[1;31mEnter your choice: \033[0m"))
selected_loc = locations[choice - 1]

result = df[df['Location'] == selected_loc].groupby('Date').size().sort_values(ascending=False).head(1)
print(f"\033[1;31mDate with most accidents in {selected_loc}:\033[0m")
print(result)
```

```python
#14 Accidents Caused by User-Selected Cause Between Two Dates

# Query with Selectable Options

# Step 1: Show all unique causes
causes = df['Cause'].dropna().unique()

print("\033[1;31mSelect a Cause of Accident:\033[0m")
for i, cause in enumerate(causes):
    print(f"{i+1}. {cause}")

# Step 2: Get user input
choice = int(input("\033[1;31mEnter your choice number: \033[0m"))
selected_cause = causes[choice - 1]

# Step 3: Get date range
start = input("\033[1;31mEnter start date (YYYY-MM-DD): \033[0m")
end = input("\033[1;31mEnter end date (YYYY-MM-DD): \033[0m")

# Step 4: Filter and display
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
mask = (df['Cause'] == selected_cause) & (df['Date'] >= start) & (df['Date'] <= end)

print(f"\n✅ \033[1;31mTotal accidents due to '{selected_cause}' between {start} and {end}: \033[0m\033[1;31m{df[mask].shape[0]}\033[0m")
```

```python
#15 Monthly Accident Trend by Selected Location (With Options)

# Ensure Date is datetime and extract Month
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.to_period('M')

# Step 1: Show all unique locations
locations = df['Location'].dropna().unique()

print("\033[1;31mSelect a Location:\033[0m")
for i, loc in enumerate(locations):
    print(f"{i+1}. {loc}")

# Step 2: Get user input
choice = int(input("\033[1;31mEnter your choice number: \033[0m"))
selected_location = locations[choice - 1]

# Step 3: Group by month
trend = df[df['Location'] == selected_location].groupby('Month').size()

# Step 4: Show results
print(f"\n📊\033[1;31m Monthly accident trend for '{selected_location}':\033[0m")
print("\033[1;33m",trend,"\033[0m")
```

```python
#16. Compare Accidents Between Two Locations Over Months

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.to_period('M')

locations = df['Location'].dropna().unique()
print("\033[1;31mSelect First Location:\033[0m")
for i, loc in enumerate(locations): print(f"{i+1}. {loc}")
loc1 = locations[int(input("\033[1;31mEnter choice number: \033[0m")) - 1]

print("\033[1;31mSelect Second Location:\033[0m")
for i, loc in enumerate(locations): print(f"{i+1}. {loc}")
loc2 = locations[int(input("\033[1;31mEnter choice number: \033[0m")) - 1]

grp = df[df['Location'].isin([loc1, loc2])].groupby(['Location', 'Month']).size().unstack(0).fillna(0)
print(f"\n\033[1;31mMonthly accident comparison between {loc1} and {loc2}:\033[0m")
print("\033[1;33m",grp,"\033[0m")
```

```python
#17. Compare Death Rate by Road Condition

grouped = df.groupby('Road Condition').agg({'No.of Deaths': 'sum', 'Accident ID': 'count'})
grouped['Death Rate'] = grouped['No.of Deaths'] / grouped['Accident ID']
print("\n💀\033[1;31m Death rate by road condition:\033[0m")
print("\033[1;33m",grouped[['Death Rate']].sort_values(by='Death Rate', ascending=False),"\033[0m")
```

```python
#18. Top 3 Locations per Month with Most Accidents (Multi-level Grouping)

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.to_period('M')

monthly_top = df.groupby(['Month', 'Location']).size().reset_index(name='Count')
monthly_top_sorted = monthly_top.sort_values(['Month', 'Count'], ascending=[True, False])

print("\n\033[1;31mTop 3 accident-prone locations each month:\033[0m")
for month in monthly_top_sorted['Month'].unique():
    print(f"\033[1;31m\nMonth: {month}\033[0m")
    print("\033[1;33m",monthly_top_sorted[monthly_top_sorted['Month'] == month].head(3),"\033[0m")
```

```python
#19. Top 5 Days With Highest Accidents Across All Locations

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
top_days = df.groupby('Date').size().sort_values(ascending=False).head(5)
print("\n\033[1;31mTop 5 days with the highest number of accidents:\033[0m")
print("\033[1;33m",top_days,"\033[0m")
```

```python
#20.Longest Accident-Free Streak per Location

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.sort_values('Date')

# Get daily records per location
locations = df['Location'].dropna().unique()
from datetime import timedelta

print("\033[1;31mLongest accident-free streaks by location:\033[0m")
for loc in locations[:10]:  # Limit output
    dates = df[df['Location'] == loc]['Date'].dropna().sort_values().unique()
    max_gap = max((dates[i+1] - dates[i]).days for i in range(len(dates)-1))
    print(f"\033[1;33m{loc}: {max_gap} days\033[0m")
```

### Data Visualization:

```python
# 1. Heatmap of Accidents by Day of Week and Hour

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Hour'] = pd.to_datetime(df['Time'], errors='coerce').dt.hour
df['Weekday'] = df['Date'].dt.day_name()

pivot = df.pivot_table(index='Weekday', columns='Hour', values='Accident ID', aggfunc='count', fill_value=0)
plt.figure(figsize=(14,6))
sns.heatmap(pivot, cmap='coolwarm', linewidths=0.5)
plt.title(' Accident Frequency Heatmap by Weekday and Hour')
plt.show()
```

```python
# 2. Monthly Trend of Accidents by Location (Line Plot with Dropdown)

import plotly.express as px

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.to_period('M').astype(str)

fig = px.line(df.groupby(['Month', 'Location']).size().reset_index(name='Accidents'),
              x='Month', y='Accidents', color='Location',
              title='Monthly Accident Trend by Location')
fig.show()
```

```python
#3. Scatter Plot: Vehicles Involved vs No. of Deaths

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Vehicles Involved', y='No.of Deaths', hue='Road Condition', size='Vehicles Involved', palette='viridis')
plt.title(' Vehicles Involved vs Deaths (Color = Road Condition)')
plt.show()
```

```python
#4. Stacked Area Chart: Monthly Accidents by Cause

df['Month'] = df['Date'].dt.to_period('M').astype(str)
monthly_cause = df.groupby(['Month', 'Cause']).size().unstack().fillna(0)

monthly_cause.plot.area(stacked=True, figsize=(14, 6), cmap='tab20c')
plt.title(" Monthly Accident Distribution by Cause")
plt.xlabel("Month")
plt.ylabel("Accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#5. Bar Plot: Accidents by Weather Condition

plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Weather Condition', order=df['Weather Condition'].value_counts().index, hue='Weather Condition', palette='mako', legend=False)
plt.title(' Accidents by Weather Condition')
plt.xlabel('Number of Accidents')
plt.ylabel('Weather Condition')
plt.show()
```
